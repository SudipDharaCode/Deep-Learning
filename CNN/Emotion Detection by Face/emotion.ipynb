```python
import cv2
from mtcnn import MTCNN
import os
import numpy as np
from keras.models import load_model
```

```python
cap = cv2.VideoCapture(0)
cap.set(3,640)
cap.set(4,480)

detector = MTCNN()

model = load_model("D:\Full_stack_ML\DL\Project\Emotion_detection\Model\emotion_detection_v2_good_cruve.h5")
emotion_dict = {0:'Angry',1:'Disgusted',2:'Fear',3:"Happy",4:"Neutral",5:'Sad',6:'Surprise'}


while True:

    ret,frame = cap.read()

    output = detector.detect_faces(frame)

    for i in output:
        for j in range(0,10):
            x,y,width,height = i['box']
        
            if i['confidence'] >= 0.80: 

                face = frame[y:y+height , x:x+width]
                face_gray = cv2.cvtColor(face,cv2.COLOR_BGR2GRAY)
                crop_face = cv2.resize(face_gray,(48,48))

                img = np.array(crop_face)
                img = img.reshape(1,48,48,1)
                img = img/255.0

                prediction = model.predict(img)[0]
                label=emotion_dict[prediction.argmax()]
                label_position = (x,y)
                cv2.rectangle(frame,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=3)
                cv2.putText(frame,label,label_position,cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
            else:
                cv2.putText(frame,'No Faces',(30,80),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
            cv2.imshow('Emotion Detector',frame)    

                # cv2.imshow("window",frame)
                # cv2.imshow("window2",frame)
                # cv2.imshow("window3",crop_face)

    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cap.release()
cv2.destroyAllWindows()
```
