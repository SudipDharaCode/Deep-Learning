```python
import cv2
from mtcnn import MTCNN

cap = cv2.VideoCapture(0)
cap.set(3,640)
cap.set(4,480)

detector = MTCNN()

while True:
    
    ret,frame = cap.read()
    out = detector.detect_faces(frame)
    
    for i in out:
        
        x,y,w,h = i['box']
        
        cv2.rectangle(frame,pt1=(x,y),pt2=(x+w,y+h),color=(0,0,255),thickness=3)
        frame[y:y+h,x:x+w] = cv2.blur(frame[y:y+h,x:x+w],(35,35))
        
    cv2.imshow("window",frame)
    
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
        
cv2.destroyAllWindows()
```
